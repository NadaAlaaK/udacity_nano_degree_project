```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from IPython.core.pylabtools import figsize
from matplotlib import rcParams

%matplotlib inline 
```

**Data preparing function**

```python
#preparing the data 
def prep_Data(file_name):
  '''Preprocessing the data .'''
  #reading data 
  df = pd.read_csv('data_cleaned_2021.csv')
  #drop index column
  df.drop("index",axis=1,inplace=True)
  #drop duplicates
  df.drop_duplicates(['Company Name'],inplace=True)
  #filtering data for data scientist roles only
  df=df[df['job_title_sim']=='data scientist']
  #sorting data 
  sorted_df=df.sort_values(by=['Avg Salary(K)','Lower Salary', 'Upper Salary'],ascending=False)
  return sorted_df

```

**Evaluate the results **

**Question 1 :** what is the most paying companies for data scientist

```python
#creating bar plot for Top 10 Companies with Highest avg Salary with the location of each company 
fig, ax = plt.subplots(nrows=1, ncols=1)
sns.barplot(x='Avg Salary(K)', y='Company Name', hue="Location", data=df[:10])
plt.title('\n Top 10 Companies with Highest avg Salary \n', size=16, color='black')

plt.show()
```

In the graph we see the highest 10  companies in the avarge salary for data scintist jobs  and the location of each company so that can help the person that seek a high salary to focus on these companies.

**Question 2** : what is the states with the max jobs for data scientist

```python
# creat barplot for the most demanding states for data scientist jobs
fig, ax = plt.subplots(nrows=1, ncols=1)
sns.barplot(x=df["Job Location"].value_counts().index[0:10], y = df["Job Location"].value_counts()[0:10])
rcParams['figure.figsize'] = 12,5
plt.title('\n Most Demanding States for Data Science Jobs \n', size=16, color='black')

plt.show()
```

In the graph we see the top 10 states that demands data scientists so if a person want a job they  have a better chance searching in those states.

**Question 3 :** what is the top 5 industries using data science

```python
# create barplot for Top 5 Industries with Most Number of Data Science Related Jobs
labels= [x for x in df["Industry"].value_counts().sort_values(ascending=False)[0:5].index] 
plt.pie(df["Industry"].value_counts().sort_values(ascending=False)[0:5],autopct='%1.1f%%',shadow=True)
plt.title('\n Top 5 Industries with Most Number of Data Science Related Jobs \n', size=16, color='black')
plt.legend(labels)
plt.axis('equal')
plt.show()
```

In the graph we see the top 5 industries that demand data scintists so if any data scientist searching for a job  they can prepare better by knowing about these industries so they can be ready working.
